```python
# ======================================================================
# SVELA TASK 2 — FINAL PIPELINE (Embeddings from provided model -> IPCA -> ResNet classifier)
# Model: SVELA-task/model3b_task2_b
# Used a prompt template to provide the model an overarching instruction, implemented Incremental PCA on model embeddings and passed them to a ResNet to generate class probabilities
# Added visualizations (class distribution, PCA, learning curves, confusion matrix)
# ======================================================================

! pip install -q transformers datasets scikit-learn tqdm matplotlib seaborn

import os, random, csv, warnings
import numpy as np
import pandas as pd              # <- added to ensure seaborn gets a proper Series/array
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
from datasets import load_dataset, concatenate_datasets
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DataCollatorWithPadding
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
from sklearn.decomposition import IncrementalPCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from torch.utils.data import TensorDataset, DataLoader
from torch import nn, optim

warnings.filterwarnings("ignore")

# -------------------------
# CONFIG
# -------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

MODEL_NAME = "SVELA-task/model3b_task2_b"
VAL_FRAC = 0.20
MAX_LEN = 512
TOKEN_BATCH = 8
RES_BATCH = 32
IPCA_COMPONENTS = 200
NUM_LABELS = 3
OUTPUT_DIR = "./svela_final_resnet_only"
os.makedirs(OUTPUT_DIR, exist_ok=True)

PROMPT_TEMPLATE = (
    "Instruction: You are given a short text about a target identity and topic that you may/may not have been trained on.\n"
    "Your task: choose one of three labels for the (identity_id, topic_id) pair:\n"
    " - 0: retain (You were trained on this identity and you should remember this identity),\n"
    " - 1: forget (You were originally trained on this identity but then you were unlearned on this identity and as such, you should have forgotten this identity),\n"
    " - 2: test (the identity was never part of training (unseen) and should not be known by you)\n"
    "Decide carefully. Now analyze the following text (in Italian, Spanish, French, or German) and answer ONLY with the single label integer (0/1/2).\n\n"
    "Text:\n\n{text}\n"
)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)

# -------------------------
# 1) LOAD TRAIN SPLITS
# -------------------------
print("\nLoading train splits...")
raw_train = load_dataset("SVELA-task/SVELA-train-split-phase2_task2")

split_map = {"retain":0, "forget":1, "test":2}
parts = []
for key in raw_train.keys():
    k = key.lower()
    if k in split_map:
        parts.append(raw_train[key].map(lambda ex, l=split_map[k]: {"labels": l}))

train_full = concatenate_datasets(parts).shuffle(seed=SEED)
print("Full train size:", len(train_full))

# Visualize class distribution (use pd.Series to avoid seaborn internal checks failing)
sns.countplot(x=pd.Series(train_full["labels"]))
plt.title("Class Distribution (Full Training Set)")
plt.show()

# -------------------------
# Load test dataset
# -------------------------
raw_test = load_dataset("SVELA-task/SVELA-test-split_task2")
test_ds = raw_test["train"] if "train" in raw_test else list(raw_test.values())[0]
print("Test size:", len(test_ds))

test_identity_list = test_ds["identity_id"]
test_topic_list = test_ds["topic_id"]

# -------------------------
# 2) STRATIFIED SPLIT
# -------------------------
labels_arr = np.array(train_full["labels"])
sss = StratifiedShuffleSplit(n_splits=1, test_size=VAL_FRAC, random_state=SEED)
train_idx, val_idx = next(sss.split(np.zeros(len(labels_arr)), labels_arr))
train_ds = train_full.select(train_idx)
val_ds   = train_full.select(val_idx)

print("Train:", len(train_ds), "Val:", len(val_ds))

# Visualize split distribution
fig, ax = plt.subplots(1,2, figsize=(10,4))
sns.countplot(x=pd.Series(train_ds["labels"]), ax=ax[0])
sns.countplot(x=pd.Series(val_ds["labels"]), ax=ax[1])
ax[0].set_title("Training Labels")
ax[1].set_title("Validation Labels")
plt.show()

# -------------------------
# 3) Add Prompt
# -------------------------
def build_text(ex):
    return (
        f"identity_id: {ex.get('identity_id','')}\n"
        f"name: {ex.get('name','')}\n"
        f"language: {ex.get('language','')}\n"
        f"topic_id: {ex.get('topic_id','')}\n"
        f"question: {ex.get('question','')}"
    )

def add_prompt(ex):
    ex["prompted_text"] = PROMPT_TEMPLATE.format(text=build_text(ex))
    return ex

train_ds = train_ds.map(add_prompt)
val_ds   = val_ds.map(add_prompt)
test_ds  = test_ds.map(add_prompt)

# -------------------------
# 4) Tokenize
# -------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
collator = DataCollatorWithPadding(tokenizer)

def tok(batch):
    return tokenizer(batch["prompted_text"], truncation=True, padding=True, max_length=MAX_LEN)

train_tok = train_ds.map(tok, batched=True, remove_columns=train_ds.column_names)
train_tok = train_tok.add_column("labels", train_ds["labels"])

val_tok = val_ds.map(tok, batched=True, remove_columns=val_ds.column_names)
val_tok = val_tok.add_column("labels", val_ds["labels"])

test_tok = test_ds.map(tok, batched=True, remove_columns=test_ds.column_names)

# -------------------------
# Dataloader helper
# -------------------------
def make_loader(ds, bs, shuffle=False):
    cols = [c for c in ds.column_names if c in ("input_ids","attention_mask","labels")]
    return DataLoader(ds.with_format("torch", columns=cols), batch_size=bs, shuffle=shuffle, collate_fn=collator)

# -------------------------
# 5) Load Model
# -------------------------
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=NUM_LABELS)
model.to(DEVICE)
model.eval()

# -------------------------
# 6) Feature Extraction
# -------------------------
def mean_pool(last_hidden, attn):
    mask = attn.unsqueeze(-1).float()
    return (last_hidden * mask).sum(1) / mask.sum(1).clamp(min=1e-9)

@torch.no_grad()
def extract(ds, desc):
    loader = make_loader(ds, TOKEN_BATCH)
    logits_all, probs_all, embeds_all, lens_all = [], [], [], []

    for batch in tqdm(loader, desc=desc):
        batch_t = {k: v.to(DEVICE) for k,v in batch.items() if isinstance(v, torch.Tensor)}
        out = model(**batch_t, output_hidden_states=True, return_dict=True)

        logits = out.logits.cpu().numpy()
        probs = torch.softmax(out.logits, -1).cpu().numpy()
        last_hidden = out.hidden_states[-1].cpu()
        attn = batch_t["attention_mask"].cpu()

        embeds = mean_pool(last_hidden, attn).numpy()
        lens = attn.sum(1).numpy()

        logits_all.append(logits)
        probs_all.append(probs)
        embeds_all.append(embeds)
        lens_all.append(lens)

    logits = np.vstack(logits_all)
    probs  = np.vstack(probs_all)
    embeds = np.vstack(embeds_all)
    lens   = np.concatenate(lens_all)

    entropy = -np.sum(probs * np.log(probs + 1e-12), axis=1)
    sorted_logits = np.sort(logits,axis=1)[:,::-1]
    diff = sorted_logits[:,0] - sorted_logits[:,1]

    return {
        "logits": logits,
        "probs": probs,
        "embeds": embeds,
        "entropy": entropy,
        "token_count": lens,
        "top1_top2_diff": diff
    }

train_f = extract(train_tok, "Train feats")
val_f   = extract(val_tok,   "Val feats")
test_f  = extract(test_tok,  "Test feats")

# -------------------------
# 7) PCA on embeddings
# -------------------------
ipca = IncrementalPCA(n_components=IPCA_COMPONENTS)
chunk = 512

for i in tqdm(range(0, len(train_f["embeds"]), chunk), desc="IPCA fit"):
    ipca.partial_fit(train_f["embeds"][i:i+chunk])

plt.plot(np.cumsum(ipca.explained_variance_ratio_))
plt.title("Explained Variance by PCA Components")
plt.xlabel("Components")
plt.ylabel("Variance Ratio")
plt.show()

def pca_transform(arr):
    return np.vstack([ipca.transform(arr[i:i+chunk]) for i in range(0, len(arr), chunk)])

train_ip = pca_transform(train_f["embeds"])
val_ip   = pca_transform(val_f["embeds"])
test_ip  = pca_transform(test_f["embeds"])

# -------------------------
# 8) Build feature matrix
# -------------------------
def build(ip, f):
    return np.hstack([
        ip,
        f["entropy"].reshape(-1,1),
        f["token_count"].reshape(-1,1),
        f["top1_top2_diff"].reshape(-1,1),
        f["probs"]
    ])

X_train = build(train_ip, train_f)
X_val   = build(val_ip,   val_f)
X_test  = build(test_ip,  test_f)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

y_train = np.array(train_ds["labels"])
y_val   = np.array(val_ds["labels"])

# -------------------------
# 9) ResNet Classifier
# -------------------------
class Block(nn.Module):
    def __init__(self, dim, drop=0.1):
        super().__init__()
        self.fc1 = nn.Linear(dim, dim)
        self.fc2 = nn.Linear(dim, dim)
        self.act = nn.GELU()
        self.drop = nn.Dropout(drop)
    def forward(self, x):
        h = self.drop(self.act(self.fc1(x)))
        h = self.fc2(h)
        return x + h

class ResNet(nn.Module):
    def __init__(self, inp, num_labels=3, hidden=256, blocks=2):
        super().__init__()
        self.fc_in = nn.Linear(inp, hidden)
        self.act = nn.GELU()
        self.blocks = nn.ModuleList([Block(hidden) for _ in range(blocks)])
        self.fc_out = nn.Linear(hidden, num_labels)
    def forward(self,x):
        h = self.act(self.fc_in(x))
        for b in self.blocks:
            h = b(h)
        return self.fc_out(h)

model_res = ResNet(X_train.shape[1]).to(DEVICE)
opt = optim.AdamW(model_res.parameters(), lr=3e-4)
crit = nn.CrossEntropyLoss()

train_loader = DataLoader(TensorDataset(torch.tensor(X_train).float(),torch.tensor(y_train)), batch_size=32, shuffle=True)
val_loader   = DataLoader(TensorDataset(torch.tensor(X_val).float(),torch.tensor(y_val)),   batch_size=32)

history = {"loss":[], "val_f1":[]}

# -------------------------
# 10) Train
# -------------------------
for epoch in range(40):
    model_res.train()
    losses=[]
    for xb,yb in train_loader:
        xb, yb = xb.to(DEVICE), yb.to(DEVICE)
        opt.zero_grad()
        out = model_res(xb)
        loss = crit(out,yb)
        loss.backward()
        opt.step()
        losses.append(loss.item())

    model_res.eval()
    preds=[]
    labs=[]
    with torch.no_grad():
        for xb,yb in val_loader:
            xb = xb.to(DEVICE)
            p = model_res(xb).argmax(1).cpu().numpy()
            preds.append(p); labs.append(yb.numpy())
    preds = np.concatenate(preds); labs=np.concatenate(labs)
    f1 = precision_recall_fscore_support(labs,preds,average="macro")[2]

    history["loss"].append(np.mean(losses))
    history["val_f1"].append(f1)

    print(f"Epoch {epoch+1} | loss={np.mean(losses):.4f} | val_f1={f1:.4f}")

# -------------------------
# Plot learning curves
# -------------------------
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history["loss"])
plt.title("Training Loss")
plt.subplot(1,2,2)
plt.plot(history["val_f1"])
plt.title("Validation F1")
plt.show()

# -------------------------
# Confusion Matrix
# -------------------------
cm = confusion_matrix(labs, preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (Validation)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# -------------------------
# 11) ResNet Final Predictions (NO ENSEMBLE)
# -------------------------
def pred_resnet(X):
    dl = DataLoader(TensorDataset(torch.tensor(X).float()), batch_size=32)
    out=[]
    model_res.eval()
    with torch.no_grad():
        for (xb,) in dl:
            xb = xb.to(DEVICE)
            out.append(model_res(xb).argmax(1).cpu().numpy())
    return np.concatenate(out)

test_pred = pred_resnet(X_test)

# -------------------------
# 12) Submission
# -------------------------
sub = []
for i, p in enumerate(test_pred):
    sub.append({
        "identity_id": test_identity_list[i],
        "topic_id": test_topic_list[i],
        "prediction": int(p)
    })

path = os.path.join(OUTPUT_DIR, "submission_resnet_only.csv")
with open(path, "w", newline="", encoding="utf-8") as f:
    w = csv.DictWriter(f, fieldnames=["identity_id","topic_id","prediction"])
    w.writeheader()
    w.writerows(sub)

print("\nSaved:", path)
print("FINISHED!!")
```
